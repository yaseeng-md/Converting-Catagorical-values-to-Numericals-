# Ordinal Encoder()

```python
oe = OrdinalEncoder(categories = [["Attrited Customer","Existing Customer"],["Blue","Silver","Gold","Platinum"]])
oe.fit(data_set[["Attrition_Flag","Card_Category"]])
```

```python
data_set[["Attrition_Flag","Card_Category"]] = oe.transform(data_set[["Attrition_Flag","Card_Category"]])
```
